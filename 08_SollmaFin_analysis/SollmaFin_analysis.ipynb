SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

Динамика привлечения пользователей стабильна и предсказуема. Это создаёт хорошую базу для сравнения с A/B-экспериментом, поскольку поведение пользователей до эксперимента не зависит от экстремальных факторов.

```python
# Уникальные пользователи на каждом этапе
funnel_events = [
    'install / open_web',
    'registration',
    'main_page',
    'onboarding_complete',
    'first_deposit',
    'asset_purchase',
    'second_deposit'
]

funnel = (
    hist_df[hist_df['event_name'].isin(funnel_events)]
    .drop_duplicates(subset=['user_id', 'event_name'])
    .groupby('event_name')['user_id']
    .nunique()
    .reindex(funnel_events)
)
```

```python
# Классическая воронка
total_users = funnel.iloc[0]
classic_funnel = (funnel / total_users * 100).round(2)

# Пошаговая воронка
step_funnel = (funnel / funnel.shift(1) * 100).round(2)

# Вывод
funnel_df = pd.DataFrame({
    'Users': funnel,
    'From Start (%)': classic_funnel,
    'Step Conversion (%)': step_funnel
})

print(funnel_df)
```

```python
# Визуализация классической воронки
plt.figure(figsize=(12, 6))
sns.barplot(x=funnel.index, y=classic_funnel, palette='Blues_d')
plt.title('Классическая воронка пользователей (%)')
plt.ylabel('Процент от установивших')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

Классическая воронка демонстрирует более плавное и последовательное снижение от ~98% до ~13%, что отражает накопительный эффект потерь пользователей

```python
# Визуализация пошаговой воронки
plt.figure(figsize=(12, 6))
sns.barplot(x=funnel.index, y=step_funnel, palette='Greens_d')
plt.title('Пошаговая конверсия пользователей (%)')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

Пошаговая конверсия показывает более резкие колебания между этапами, с пиком на onboarding_complete (~ 97%) и существенными падениями на first_deposit (~ 79%) и особенно на second_purchase (~ 55%)

**Промежуточный вывод**

Наибольшие потери пользователей происходят на этапах "first_deposit" и "second_purchase", что указывает на необходимость оптимизации процессов монетизации и удержания после первой покупки.

Данные свидетельствуют о том, что продукт успешно привлекает и вовлекает пользователей на начальных этапах, но требует доработки стратегии конверсии в платящих клиентов и их долгосрочного удержания.

```python
# Отфильтруем только покупки активов
asset_df = hist_df[hist_df['event_name'] == 'asset_purchase'].copy()

# Найдём первую покупку актива для каждого пользователя
first_asset = asset_df.sort_values('event_ts').groupby('user_id').first().reset_index()

# Оставим только user_id и risk_level первой покупки
first_asset = first_asset[['user_id', 'risk_level']]
```

```python
# Отфильтруем пользователей, совершивших второй депозит
second_deposit_users = hist_df[hist_df['event_name'] == 'second_deposit']['user_id'].unique()

# Добавим бинарный флаг во второй депозит
first_asset['second_deposit'] = first_asset['user_id'].isin(second_deposit_users)
```

```python
# Группировка и расчёт доли пользователей, сделавших второй депозит
conversion_by_risk = (
    first_asset.groupby('risk_level')['second_deposit']
    .agg(['count', 'sum'])
    .rename(columns={'count': 'users', 'sum': 'converted'})
)

conversion_by_risk['conversion_rate'] = (conversion_by_risk['converted'] / conversion_by_risk['users'] * 100).round(2)
display(conversion_by_risk)
```

Это свидетельствует о том, что новая фича, вероятно, улучшает пользовательский путь и повышает вовлечённость на ключевых шагах, особенно на этапах, связанных с оплатой. Однако для окончательных выводов необходим статистический анализ значимости различий.

```python
# Фильтруем только пользователей, купивших актив
df_risk = abt_df[(abt_df['event_name'] == 'asset_purchase') & (abt_df['risk_level'].notna())]

# Получаем первую покупку актива на пользователя
risk_first_purchase = (
    df_risk.sort_values('event_ts')
    .drop_duplicates('user_id')
    [['user_id', 'group', 'risk_level']]
)

# Добавляем информацию: был ли второй депозит
second_deposit_users = abt_df[abt_df['event_name'] == 'second_deposit']['user_id'].unique()
risk_first_purchase['second_deposit'] = risk_first_purchase['user_id'].isin(second_deposit_users)

# Группировка по группе и уровню риска
summary = (
    risk_first_purchase
    .groupby(['group', 'risk_level'])['second_deposit']
    .agg(['sum', 'count'])
    .reset_index()
    .rename(columns={'sum': 'converted', 'count': 'total'})
)
summary['conversion_rate'] = (summary['converted'] / summary['total'] * 100).round(2)

# Вывод общей таблицы
print("Конверсии по группам и уровням риска:")
display(summary)
```

Общая структура гипотез для каждой подгруппы:
- H₀ (нулевая гипотеза): Конверсия во второй депозит одинакова в тестовой и контрольной группе.
- H₁ (альтернативная гипотеза): Конверсии в группах различаются.

```python
# Проверка статистической значимости между группами
print("Гипотезы:\n")
print("H₀: Конверсии равны между группами (test = control)")
print("H₁: Конверсии различаются между группами (test ≠ control)\n")

print("Результаты Z-теста:\n")
risk_levels = summary['risk_level'].unique()
for risk in risk_levels:
    test = summary[(summary['group'] == 'test') & (summary['risk_level'] == risk)]
    control = summary[(summary['group'] == 'control') & (summary['risk_level'] == risk)]

    count = [int(test['converted']), int(control['converted'])]
    nobs = [int(test['total']), int(control['total'])]

    zstat, pval = proportions_ztest(count, nobs)
    print(f"• {risk.upper()} — p-value: {pval:.5f}, z-stat: {zstat:.2f}")
```

**Промежуточный вывод**

Анализ поведения пользователей, совершивших первую покупку актива, показывает:

- Тестовая группа демонстрирует более высокую конверсию во второй депозит во всех категориях риска (high, medium, low), однако статистическая значимость различий варьируется в зависимости от уровня риска.

- ***High-risk***
  - Конверсия выросла с 27.66% до 46.00%.
  - p-value < 0.001 — разница статистически значима.
  - Наиболее заметный прирост — новая фича особенно эффективна для пользователей, склонных к высоким рискам.
- ***Low-risk***
  - Конверсия увеличилась почти на 8 п.п. — с 31.48% до 39.47%.
  - p-value ≈ 0.039 — разница также статистически значима, но в меньшей степени.
- ***Medium-risk***
  - Конверсия выросла с 44.54% до 47.04%.
  - p-value ≈ 0.36 — разница не является статистически значимой.
  
Новая версия онбординга оказывает наиболее выраженное положительное влияние на пользователей, выбирающих высокорисковые и низкорисковые активы, увеличивая их склонность к повторным инвестициям. Это может указывать на более эффективную адаптацию и повышение доверия к платформе, особенно у пользователей с полярными типами риска.

```python
# Функции для тестов 
def run_ttest(group_a, group_b):
    """Выполняет t-тест и возвращает статистику и p-value"""
    t_stat, p_val = ttest_ind(group_a, group_b, equal_var=False)
    return p_val

def run_ztest(successes, totals):
    """Выполняет Z-тест для пропорций и возвращает p-value"""
    stat, p_val = proportions_ztest(successes, totals)
    return p_val

def mean_diff(a, b):
    return b.mean() - a.mean()

def rel_diff(a, b):
    return (b.mean() / a.mean() - 1) * 100
```

```python
# Данные пользователей и событий
users = abt_df[['user_id', 'group']].drop_duplicates()
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])]
registrations = abt_df[abt_df['event_name'] == 'registration'][['user_id', 'group']]
```

```python
# Ключевая метрика
total_deposits = deposits.groupby('user_id')['amount'].sum().reset_index()
key_metric = users.merge(total_deposits, on='user_id', how='left').fillna(0)
group_key = [key_metric[key_metric['group'] == g]['amount'] for g in ['control', 'test']]
```

```python
# Барьерная метрика
first_deposit_users = abt_df[abt_df['event_name'] == 'first_deposit']['user_id'].unique()
registrations['converted'] = registrations['user_id'].isin(first_deposit_users)
barrier_counts = registrations.groupby('group')['converted'].agg([np.sum, 'count'])
```

```python
# Вспомогательная метрика 1
fd_users = abt_df[abt_df['event_name'] == 'first_deposit'][['user_id', 'group']]
sd_users = abt_df[abt_df['event_name'] == 'second_deposit']['user_id'].unique()
fd_users['converted'] = fd_users['user_id'].isin(sd_users)
assist_counts = fd_users.groupby('group')['converted'].agg([np.sum, 'count'])

# Вспомогательная метрика 2
deposit_users = total_deposits[total_deposits['amount'] > 0].merge(users, on='user_id', how='left')
group_assist2 = [deposit_users[deposit_users['group'] == g]['amount'] for g in ['control', 'test']]
```

```python
# Сбор результатов

summary = {
    'Ключевая метрика (сред. сумма всех депозитов)': {
        'control': group_key[0].mean(),
        'test': group_key[1].mean(),
        'abs_diff': mean_diff(group_key[0], group_key[1]),
        'rel_diff_%': rel_diff(group_key[0], group_key[1]),
        'p_value': run_ttest(group_key[0], group_key[1])
    },
    'Барьерная метрика (регистрация → депозит 1)': {
        'control': barrier_counts.loc['control', 'sum'] / barrier_counts.loc['control', 'count'],
        'test': barrier_counts.loc['test', 'sum'] / barrier_counts.loc['test', 'count'],
        'abs_diff': (
            barrier_counts.loc['test', 'sum'] / barrier_counts.loc['test', 'count'] -
            barrier_counts.loc['control', 'sum'] / barrier_counts.loc['control', 'count']
        ),
        'rel_diff_%': (
            barrier_counts.loc['test', 'sum'] / barrier_counts.loc['test', 'count'] /
            (barrier_counts.loc['control', 'sum'] / barrier_counts.loc['control', 'count']) - 1
        ) * 100,
        'p_value': run_ztest(barrier_counts['sum'], barrier_counts['count'])
    },
    'Вспомогательная метрика 1 (депозит 1 → депозит 2)': {
        'control': assist_counts.loc['control', 'sum'] / assist_counts.loc['control', 'count'],
        'test': assist_counts.loc['test', 'sum'] / assist_counts.loc['test', 'count'],
        'abs_diff': (
            assist_counts.loc['test', 'sum'] / assist_counts.loc['test', 'count'] -
            assist_counts.loc['control', 'sum'] / assist_counts.loc['control', 'count']
        ),
        'rel_diff_%': (
            assist_counts.loc['test', 'sum'] / assist_counts.loc['test', 'count'] /
            (assist_counts.loc['control', 'sum'] / assist_counts.loc['control', 'count']) - 1
        ) * 100,
        'p_value': run_ztest(assist_counts['sum'], assist_counts['count'])
    },
    'Вспомогательная метрика 2 (сред. сумма депозитов у тех, у кого они есть)': {
        'control': group_assist2[0].mean(),
        'test': group_assist2[1].mean(),
        'abs_diff': mean_diff(group_assist2[0], group_assist2[1]),
        'rel_diff_%': rel_diff(group_assist2[0], group_assist2[1]),
        'p_value': run_ttest(group_assist2[0], group_assist2[1])
    }
}

# Вывод результата
pd.DataFrame(summary).T
```

```python
# Приводим event_ts к дате (без времени)
abt_df['event_date'] = pd.to_datetime(abt_df['event_ts']).dt.date

# Суммарные депозиты по пользователям по дням
daily_deposits = (
    abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])]
    .groupby(['event_date', 'user_id', 'group'])['amount']
    .sum()
    .reset_index()
)

# Кумулятивная сумма по каждому пользователю
daily_deposits['cum_sum'] = (
    daily_deposits
    .sort_values(['group', 'user_id', 'event_date'])
    .groupby(['group', 'user_id'])['amount']
    .cumsum()
)

# Средняя кумулятивная сумма по дням и группам
cumulative = (
    daily_deposits
    .groupby(['event_date', 'group'])['cum_sum']
    .mean()
    .reset_index()
)

# Визуализация
plt.figure(figsize=(10, 5))
sns.lineplot(data=cumulative, x='event_date', y='cum_sum', hue='group')
plt.title('Кумулятивная средняя сумма депозитов по группам')
plt.xlabel('Дата')
plt.ylabel('Средняя сумма депозитов')
plt.grid(True)
plt.tight_layout()
plt.show()
```

График демонстрирует кумулятивную среднюю сумму депозитов, где тестовая группа (оранжевая линия) показывает заметно лучшие результаты, особенно с середины июня. К концу периода тестовая группа достигла ~380 единиц против ~290 в контрольной группе.

```python
# Инициализируем список
p_values = []

# Уникальные даты
dates = sorted(daily_deposits['event_date'].unique())

for date in dates:
    # Данные до и включая текущую дату
    snapshot = daily_deposits[daily_deposits['event_date'] <= date]

    # Сумма депозитов на пользователя
    total_by_user = snapshot.groupby(['user_id', 'group'])['amount'].sum().reset_index()

    # Делим по группам
    group_control = total_by_user[total_by_user['group'] == 'control']['amount']
    group_test = total_by_user[total_by_user['group'] == 'test']['amount']

    # Только если в обеих группах есть хотя бы по 2 наблюдения
    if len(group_control) >= 2 and len(group_test) >= 2:
        try:
            _, p_val = ttest_ind(group_control, group_test, equal_var=False)
            if np.isfinite(p_val):
                p_values.append({'date': date, 'p_value': p_val})
        except Exception as e:
            # Пропускаем проблемные итерации
            continue

# В датафрейм и очистка
pval_df = pd.DataFrame(p_values)

# Построим график
plt.figure(figsize=(10, 5))
sns.lineplot(data=pval_df, x='date', y='p_value')
plt.axhline(0.05, color='red', linestyle='--', label='α = 0.05')
plt.title('Динамика p-value по дням')
plt.xlabel('Дата')
plt.ylabel('p-value')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

График показывает динамику p-value теста, где видно, что в период с 8 по 12 июня наблюдались статистически значимые различия между группами (p-value < 0.05). После 13 июня эффект стал менее выражен, а к концу периода различия практически исчезли.

**Промежуточный вывод**

- Основная метрика — средняя сумма депозитов на пользователя — демонстрирует рост в тестовой группе, однако различия не достигают статистической значимости по итогам всего периода.

- Вспомогательные метрики:

  - Средний чек среди пользователей, совершивших хотя бы один депозит, также заметно увеличился.
  - Барьерная метрика (регистрация → первый депозит) показывает незначительное снижение в тестовой группе, но без критического отклонения.

- График динамики p-value:

Новая фича показывает потенциал в части удержания и повышения вовлечённости через повторные депозиты и рост среднего чека. Однако эффект не является устойчивым на всём горизонте — возможно, требуется доработка механики или дополнительных триггеров для пользователей.

### 4. Анализ изменений суммы депозитов на платящего пользователя 

```python
# Фильтруем только события депозитов
deposits = abt_df[abt_df['event_name'].isin(['first_deposit', 'second_deposit'])]

# Считаем сумму депозитов по пользователям
deposits_grouped = deposits.groupby(['user_id', 'group'])['amount'].sum().reset_index()

# Оставляем только платящих пользователей
paying = deposits_grouped[deposits_grouped['amount'] > 0]
```

```python
# Разделяем по группам
control = paying[paying['group'] == 'control']['amount'].values
test = paying[paying['group'] == 'test']['amount'].values

# Настройки бутстрапа
n_iterations = 5000
np.random.seed(42)

results = {}

for q in [25, 50, 75]:
    boot_diffs = []
    for _ in range(n_iterations):
        boot_c = np.random.choice(control, size=len(control), replace=True)
        boot_t = np.random.choice(test, size=len(test), replace=True)
        
        control_q = np.percentile(boot_c, q)
        test_q = np.percentile(boot_t, q)
        boot_diffs.append(test_q - control_q)
    
    boot_diffs = np.array(boot_diffs)
    ci = np.percentile(boot_diffs, [2.5, 97.5])
    
    results[q] = {
        'control_percentile': np.percentile(control, q),
        'test_percentile': np.percentile(test, q),
        'diff_mean': np.mean(boot_diffs),
        'ci_low': ci[0],
        'ci_high': ci[1],
        'boot_diffs': boot_diffs
    }
```

```python
# Вывод результатов
for q in [25, 50, 75]:
    res = results[q]
    print(f"{q}-й перцентиль:")
    print(f"  Control: {res['control_percentile']:.2f}")
    print(f"  Test:    {res['test_percentile']:.2f}")
    print(f"  Разница: {res['diff_mean']:.2f}")
    print(f"  95% CI:  [{res['ci_low']:.2f}, {res['ci_high']:.2f}]\n")
```

```python
# Визуализация
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for i, q in enumerate([25, 50, 75]):
    sns.histplot(results[q]['boot_diffs'], kde=True, ax=axes[i])
    axes[i].axvline(0, color='red', linestyle='--')
    axes[i].set_title(f"Bootstrap распределение разности {q}-го перцентиля")
    axes[i].set_xlabel('Разность (test - control)')
    axes[i].set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

- А большая часть — меньше или вовсе прекратила депозиты, что снижает медиану.

**Промежуточный вывод**

По результатам теста видно, что тестовая группа показывает неоднородное поведение:

- Общая сумма депозитов на пользователя в среднем растёт, но статистическая значимость достигается только на отдельных интервалах.

- Конверсия в повторный депозит и средний чек активных пользователей увеличиваются — это положительный сигнал.

- Однако наблюдается снижение медианы и нижнего квартиля:
  - 25-й перцентиль снизился с 95 до 48
  - Медиана с 118 до 94
  - 75-й перцентиль, напротив, вырос с 159 до 179.5

Графики распределения и бутстрап-перцентили показывают, что:
- В тестовой группе появилась прослойка пользователей с меньшими депозитами — либо они стали осторожнее, либо совершили только минимальный вклад.
- Одновременно усилился правый хвост: некоторые пользователи стали вносить гораздо больше, чем в контрольной группе.

Медиана снижается не потому, что всё стало хуже, а потому, что поведение стало более полярным. Часть пользователей «платит меньше или уходит», часть — платит намного больше.

Фича усилила сегментацию аудитории. Возможно, она лучше работает для опытных или вовлечённых пользователей, но добавляет недоверие у массового сегмента. Это открывает потенциал для таргетирования и персонализации.
